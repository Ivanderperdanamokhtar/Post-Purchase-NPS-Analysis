**Post Purchase NPS Analysis** merupakan metode evaluasi yang menggunakan Net Promoter Score (NPS) untuk menilai kepuasan dan loyalitas pelanggan setelah mereka membeli produk atau menggunakan layanan. Analisis ini bertujuan untuk memahami pengalaman pelanggan pasca pembelian, mengidentifikasi aspek yang mereka sukai, serta menemukan area yang perlu diperbaiki guna meningkatkan hubungan jangka panjang antara pelanggan dan perusahaan.

**Komponen Utama dalam Post Purchase NPS Analysis**

1 Net Promoter Score (NPS)


* NPS adalah metrik yang digunakan untuk mengukur kesediaan pelanggan merekomendasikan produk atau layanan kepada orang lain.
* Skor dihitung berdasarkan pertanyaan utama:
"Seberapa besar kemungkinan Anda merekomendasikan [produk/perusahaan] kepada teman atau kolega?"
* Pelanggan memberikan skor dari 0 hingga 10, yang kemudian dikategorikan sebagai:
** Promoter (9-10): Pelanggan sangat puas dan cenderung merekomendasikan.
** Passive (7-8): Pelanggan puas tetapi tidak terlalu antusias.
** Detractor (0-6): Pelanggan tidak puas dan mungkin memberikan ulasan negatif.


2 Waktu Pelaksanaan
* Survei dilakukan setelah pembelian, biasanya dalam waktu beberapa hari hingga minggu setelah pelanggan menggunakan produk atau layanan, untuk memastikan tanggapan mencerminkan pengalaman nyata mereka.

3 Tujuan Utama
* Mengidentifikasi tingkat kepuasan pelanggan setelah pembelian.
Memahami apa yang disukai dan tidak disukai pelanggan tentang produk atau pengalaman pembelian mereka.
Mengumpulkan wawasan untuk meningkatkan produk, layanan, atau proses pembelian.


4 Analisis Hasil
* Skor NPS dihitung menggunakan formula: NPS = %Promoter − %Detractor
* Feedback kualitatif yang diberikan pelanggan (jika disertai pertanyaan tambahan) dianalisis untuk memahami penyebab skor tertentu.
* Data ini digunakan untuk mengidentifikasi area yang memerlukan perbaikan atau untuk menyoroti kekuatan perusahaan.


**Manfaat Post Purchase NPS Analysis**
1. Memberikan wawasan tentang pengalaman pelanggan setelah pembelian.
2. Membantu perusahaan mengidentifikasi pelanggan setia dan memperbaiki pengalaman pelanggan yang tidak puas.
3. Menyediakan data untuk strategi pemasaran berbasis pelanggan, seperti menciptakan promosi yang ditargetkan untuk Promoters.
4. Memperbaiki proses penjualan, layanan, atau produk berdasarkan masukan pelanggan.

Dengan memahami Post Purchase NPS Analysis, perusahaan dapat menciptakan pengalaman pelanggan yang lebih baik dan meningkatkan loyalitas pelanggan jangka panjang.


**Study Case**

Bayangkan sebuah e-commerce fashion bernama StyleHub, yang menjual pakaian trendi untuk anak muda. Selama setahun terakhir, StyleHub mengalami peningkatan jumlah pelanggan baru berkat strategi pemasaran yang kreatif. Tapi, ada masalah besar: hanya 15% pelanggan yang melakukan pembelian kedua.

Tim Data Analytics StyleHub memutuskan untuk melakukan survei Post Purchase NPS Analysis. Survei ini dikirim melalui email kepada pelanggan 7 hari setelah mereka menerima produk, dengan pertanyaan utama:

"Seberapa besar kemungkinan Anda merekomendasikan StyleHub kepada teman atau keluarga? (0-10)"

Selain itu, mereka juga menambahkan pertanyaan terbuka:
1. Apa yang Anda sukai dari pengalaman Anda?
2. Apa yang perlu kami tingkatkan?

**Step 04 : Data Findings**


```python
import pandas as pd

# Memuat dataset yang disediakan oleh pengguna
file_path = "/content/PostPurchaseNPSDataset 1.csv"
user_dataset = pd.read_csv(file_path)

# Memfilter dataset untuk Detractors (NPS_Score 0-6)
detractors = user_dataset[user_dataset['NPS_Score'] <= 6]

# Menganalisis umpan balik untuk mengidentifikasi alasan paling umum dari skor NPS rendah
feedback_summary = detractors['Feedback'].value_counts()

# Menampilkan alasan utama
feedback_summary.head(10)
```

**1. Apa penyebab utama pelanggan memberikan skor NPS rendah (Detractor)?**

```python
import matplotlib.pyplot as plt

# Memilih alasan utama untuk skor NPS rendah
top_reasons = feedback_summary.head(8)

# Membuat diagram batang dengan nilai yang dilabeli pada setiap batang
plt.figure(figsize=(10, 6))
bars = plt.bar(top_reasons.index, top_reasons.values, color='#98FB98', edgecolor='black', alpha=0.8)

# Menambahkan nilai di atas setiap batang
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + (bar.get_height() * 0.05),
             str(bar.get_height()), ha='center', fontsize=12)

# Menambahkan judul dan label
plt.title("Penyebab Utama Skor NPS Rendah (Detractor)", fontsize=16, pad=20)
plt.xlabel("Penyebab", fontsize=14, labelpad=10)
plt.ylabel("Jumlah Kasus", fontsize=14, labelpad=10)
plt.xticks(rotation=45, ha='right', fontsize=12)

# Mengatur batas atas sumbu y agar batang tertinggi melewati garis bingkai
plt.ylim(0, max(top_reasons.values) * 1.2)

plt.tight_layout()

# Menampilkan diagram
plt.show()
```

**2. Apakah ada pola tertentu pada pelanggan yang puas (Promoter) dibandingkan yang tidak puas?**

```python
# Memfilter dataset untuk Promoters (NPS_Score 9-10) dan Detractors (NPS_Score 0-6)
promoters = user_dataset[user_dataset['NPS_Score'] >= 9]
detractors = user_dataset[user_dataset['NPS_Score'] <= 6]

# Menganalisis pola potensial dengan membandingkan statistik ringkasan dan umpan balik
promoters_summary = promoters.describe(include='all')
detractors_summary = detractors.describe(include='all')

# Membandingkan atribut utama antara Promoters dan Detractors
feedback_promoters = promoters['Feedback'].value_counts()
feedback_detractors = detractors['Feedback'].value_counts()

promoters_summary, feedback_promoters.head(5), feedback_detractors.head(5)
```

```python
# Membuat diagram batang horizontal berdampingan untuk feedback utama dari Promoters dan Detractors
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# Visualisasi feedback Promoters
bars_promoters = axes[0].barh(feedback_promoters.index[:5], feedback_promoters.values[:5], color='green', edgecolor='black')
axes[0].set_title("Feedback Utama dari Promoters", fontsize=14)
axes[0].set_xlabel("Jumlah Kasus", fontsize=12)
axes[0].set_ylabel("Feedback", fontsize=12)
for bar in bars_promoters:
    axes[0].text(bar.get_width() + 0.5, bar.get_y() + bar.get_height() / 2, str(int(bar.get_width())), va='center', fontsize=12)

# Visualisasi feedback Detractors
bars_detractors = axes[1].barh(feedback_detractors.index[:5], feedback_detractors.values[:5], color='red', edgecolor='black')
axes[1].set_title("Feedback Utama dari Detractors", fontsize=14)
axes[1].set_xlabel("Jumlah Kasus", fontsize=12)
for bar in bars_detractors:
    axes[1].text(bar.get_width() + 0.5, bar.get_y() + bar.get_height() / 2, str(int(bar.get_width())), va='center', fontsize=12)

# Menyesuaikan tata letak dan menampilkan diagram
plt.tight_layout()
plt.show()
```



---

3. Pola Utama: Perbedaan utama terletak pada kecepatan pengiriman dan kesesuaian produk dengan ekspektasi pelanggan. Meningkatkan kedua aspek ini dapat meningkatkan jumlah Promoters dan mengurangi Detractors.

**3. Faktor apa yang paling sering disebutkan dalam feedback pelanggan terkait ketidakpuasan (warna produk, pengiriman, ukuran, dll.)?**

```python
from collections import Counter

# Ekstrak semua umpan balik dari detractors
detractor_feedback = detractors['Feedback']

# Mendefinisikan faktor potensial
factors = ["warna", "pengiriman", "ukuran", "kemasan", "pelayanan", "cacat", "pengalaman"]

# Menghitung kemunculan setiap faktor dalam umpan balik
factor_counts = Counter()
for feedback in detractor_feedback:
    for factor in factors:
        if factor in feedback.lower():
            factor_counts[factor] += 1

# Mengonversi ke daftar yang diurutkan
sorted_factor_counts = sorted(factor_counts.items(), key=lambda x: x[1], reverse=True)

# Menampilkan hasil
sorted_factor_counts
```

```python
# Menyiapkan data untuk visualisasi
factors, counts = zip(*sorted_factor_counts)

# Membuat diagram batang horizontal untuk faktor yang disebutkan dalam feedback
plt.figure(figsize=(10, 6))
bars = plt.barh(factors, counts, color='red', edgecolor='black')

# Menambahkan nilai pada setiap batang
for bar in bars:
    plt.text(bar.get_width() + 0.5, bar.get_y() + bar.get_height() / 2,
             str(int(bar.get_width())), va='center', fontsize=12)

# Menambahkan judul dan label
plt.title("Faktor yang Paling Sering Disebutkan dalam Feedback Ketidakpuasan", fontsize=16)
plt.xlabel("Jumlah Sebutan", fontsize=14)
plt.ylabel("Faktor", fontsize=14)
plt.tight_layout()

# Menampilkan diagram
plt.show()
```

**Tahap Produk**

Masalah yang Ditemukan:

Optimalkan Mitra Logistik:
